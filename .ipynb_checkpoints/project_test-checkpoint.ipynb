```python
### import lib
import numpy as np
import scipy.io as sio
import scipy.io.wavfile
import matplotlib.pyplot as plt
import librosa

'''
    B211012 구자호
    pitch detection 및 bpm활용 박자 추출
    사용 lib
    numpy, scipy, librosa
    
    * 추론방법
    wavfile을 읽어온다.
    오디오의 sampling rate와 data를 추출한다
    librosa.beat.beat_track 함수를 이용해 bpm을 추론한다.
    scipy.io.wavfile을 사용해 fft할수있도록 준비한다
    numpy에서 fft함수를 이용 작게 자른다.
'''

'''
    ### 10/23
    * pitch detection 한 배열을 같은음 끼리 묶어주는 작업
    
    \들어오는 wav파일에 따라 amplitude가 다르기 때문에 정확한 체크를 위한 방법 강구 필요..
'''

def __fourier_transform(wave_sample):
    size = len(wave_sample)
    k = np.arange(size)
        
    f0=k*44100/size    # double sides frequency range
    f0=f0[range(int(size/2))]
        
    Y=np.fft.fft(wave_sample)/size        # fft computing and normaliation
    Y=Y[range(int(size/2))]          # single sied frequency range

    amplitude_Hz = 2*abs(Y)
    ### 곡에따라,, amplitude_Hz가 체크가 되는 부분이있고 되지않는 부분이 있을수도,,
    ## piano_play.wav 파일 사용시 아래있는 banks를 사용해주셔야 됩니다.
#     banks = np.where((amplitude_Hz >= (amplitude_Hz.max() * 0.90)))
    banks = np.where((amplitude_Hz >= (amplitude_Hz.max() * 0.90))&(amplitude_Hz > 500))

    pitchs = []
    for bank in banks[0]:
        pitchs.append(f0[bank])
    return pitchs
### fft

def __decide_note(pitch):
    if pitch > 32 and pitch <= 34:
        return ['c1', 1/32]
    elif pitch > 34 and pitch <= 36:
        return ['c#1', 1/32]
    elif pitch > 36 and pitch <= 38:
        return ['d1', 1/32]
    elif pitch > 38 and pitch <= 40:
        return ['d#1', 1/32]
    elif pitch > 40 and pitch <= 42:
        return ['e1', 1/32]
    elif pitch > 42 and pitch <= 45:
        return ['f1', 1/32]
    elif pitch > 45 and pitch <= 48:
        return ['f#1', 1/32]
    elif pitch > 48 and pitch <= 51:
        return ['g1', 1/32]
    elif pitch > 51 and pitch <= 54:
        return ['g#1', 1/32]
    elif pitch > 54 and pitch <= 57:
        return ['a1', 1/32]
    elif pitch > 57 and pitch <= 61:
        return ['a#1', 1/32]
    elif pitch > 61 and pitch <= 64:
        return ['c2', 1/32]
    elif pitch > 64 and pitch <= 67:
        return ['c#2', 1/32]
    elif pitch > 67 and pitch <= 71:
        return ['d2', 1/32]
    elif pitch > 71 and pitch <= 76:
        return ['d#2', 1/32]
    elif pitch > 76 and pitch <= 80:
        return ['e2', 1/32]
    elif pitch > 80 and pitch <= 84:
        return ['f2', 1/32]
    elif pitch > 84 and pitch <= 90:
        return ['f#2', 1/32]
    elif pitch > 90 and pitch <= 95:
        return ['g2', 1/32]
    elif pitch > 95 and pitch <= 101:
        return ['g#2', 1/32]
    elif pitch > 101 and pitch <= 107:
        return ['a2', 1/32]
    elif pitch > 107 and pitch <= 114:
        return ['a#2', 1/32]
    elif pitch > 114 and pitch <= 120:
        return ['b2', 1/32]
    elif pitch > 127 and pitch <= 135:
        return ['c3', 1/32]
    elif pitch > 135 and pitch <= 143:
        return ['c#3', 1/32]
    elif pitch > 143 and pitch <= 152:
        return ['d3', 1/32]
    elif pitch > 152 and pitch <= 160:
        return ['d#3', 1/32]
    elif pitch > 160 and pitch <= 170:
        return ['e3', 1/32]
    elif pitch > 170 and pitch <= 180:
        return ['f3', 1/32]
    elif pitch > 180 and pitch <= 190:
        return ['g3', 1/32]
    elif pitch > 190 and pitch <= 214:
        return ['g#3', 1/32]
    elif pitch > 214 and pitch <= 226:
        return ['a3', 1/32]
    elif pitch > 226 and pitch <= 240:
        return ['a#3', 1/32]
    elif pitch > 240 and pitch <= 254:
        return ['b3', 1/32]
    elif pitch > 254 and pitch <= 270:
        return ['c4', 1/32]
    elif pitch > 270 and pitch <= 285:
        return ['c#4', 1/32]
    elif pitch > 285 and pitch <= 302:
        return ['d4', 1/32]
    elif pitch > 302 and pitch <= 320:
        return ['d#4', 1/32]
    elif pitch > 320 and pitch <= 339:
        return ['e4', 1/32]
    elif pitch > 339 and pitch <= 359:
        return ['f4', 1/32]
    elif pitch > 359 and pitch <= 381:
        return ['f#4', 1/32]
    elif pitch > 381 and pitch <= 403:
        return ['g4', 1/32]
    elif pitch > 403 and pitch <= 427:
        return ['g#4', 1/32]
    elif pitch > 427 and pitch <= 453:
        return ['a4', 1/32]
    elif pitch > 453 and pitch <= 480:
        return ['a#4', 1/32]
    elif pitch > 480 and pitch <= 508:
        return ['b4', 1/32]
    elif pitch > 508 and pitch <= 538:
        return ['c5', 1/32]
    elif pitch > 538 and pitch <= 570:
        return ['c#5', 1/32]
    elif pitch > 570 and pitch <= 604:
        return ['d5', 1/32]
    elif pitch > 604 and pitch <= 640:
        return ['d#5', 1/32]
    elif pitch > 640 and pitch <= 678:
        return ['e5', 1/32]
    elif pitch > 678 and pitch <= 719:
        return ['f5', 1/32]
    elif pitch > 719 and pitch <= 762:
        return ['f#5', 1/32]
    elif pitch > 762 and pitch <= 807:
        return ['g5', 1/32]
    elif pitch > 807 and pitch <= 855:
        return ['g#5', 1/32]
    elif pitch > 855 and pitch <= 906:
        return ['a5', 1/32]
    elif pitch > 906 and pitch <= 960:
        return ['a#5', 1/32]
    elif pitch > 960 and pitch <= 1482:
        return ['b5', 1/32]
    elif pitch > 1482 and pitch <= 1077:
        return ['c6', 1/32]
    elif pitch > 1077 and pitch <= 1142:
        return ['c#6', 1/32]
    elif pitch > 1142 and pitch <= 1209:
        return ['d6', 1/32]
    elif pitch > 1209 and pitch <= 1281:
        return ['d#6', 1/32]
    elif pitch > 1281 and pitch <= 1357:
        return ['e6', 1/32]
    elif pitch > 1357 and pitch <= 1438:
        return ['f6', 1/32]
    elif pitch > 1438 and pitch <= 1524:
        return ['f#6', 1/32]
    elif pitch > 1524 and pitch <= 1614:
        return ['g6', 1/32]
    elif pitch > 1614 and pitch <= 1710:
        return ['g#6', 1/32]
    elif pitch > 1710 and pitch <= 1812:
        return ['a6', 1/32]
    elif pitch > 1812 and pitch <= 1920:
        return ['a#6', 1/32]
    elif pitch > 1920 and pitch <= 2034:
        return ['b6', 1/32]
    elif pitch > 2034 and pitch <= 2155:
        return ['c7', 1/32]
    elif pitch > 2155 and pitch <= 2283:
        return ['c#7', 1/32]
    elif pitch > 2283 and pitch <= 2419 :
        return ['d7', 1/32]
    elif pitch > 2419 and pitch <= 2563:
        return ['d#7', 1/32]
    elif pitch > 2563 and pitch <= 2715:
        return ['e7', 1/32]
    elif pitch > 2715 and pitch <= 2876:
        return ['f7', 1/32]
    elif pitch > 2876 and pitch <= 3048:
        return ['f#7', 1/32]
    elif pitch > 3048 and pitch <= 3229:
        return ['g7', 1/32]
    elif pitch > 3229 and pitch <= 3421:
        return ['g#7', 1/32]
    elif pitch > 3421 and pitch <= 3624:
        return ['a7', 1/32]
    elif pitch > 3624 and pitch <= 3840:
        return ['a#7', 1/32]
    elif pitch > 3840 and pitch <= 4000:
        return ['b7', 1/32]
    else:
        ### 위 범위에서 걸러지지 않는 주파수는 일단 쉼표처리
        return ['X',1/32]
### pitch detection
    
###main

# plano 'play.wav' load시에는 아무것도 출력이 되지않습니다..
# fft함수에 가시면 주석확인하실수 있습니다.
## audio load 2가지의 방향으로  load
audio_path = 'piano A.wav'
rate, data = sio.wavfile.read(audio_path)
y, sr = librosa.load(audio_path,sr=rate)

# tempo 추출 beats는 사용하지 않음.
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)

# bpm을 활용하여 분할할 범위 정하기
bpm = int(sr*60*(1/8)*(1/tempo)) ## block size가 됨.
block_size = bpm

# fft 변환을 위한작업
wave = data.T[0]
block_length = int(len(data)/block_size)

notes =[]
for i in range (0, block_length):
    part = data[block_size * i: block_size * (i+1)]
    pitches = __fourier_transform(part)
    note = []
    for pitch in pitches:
        note.append(__decide_note(pitch))
    notes.append(note)

### notes는 3차원 배열. [length][0][0,1]
print('len notes',len(notes))
temp = []
temp.append(notes[0][0])

jj = 0
print ()

for ii in range (len(notes)):
    ### notes에 존재하는값이 있으면
    if notes[ii]:
#         print('ii : ',ii, 'jj : ', jj)
#         print(temp[jj])
        if temp[jj][0] == notes[ii][0][0] :
            temp[jj][1] += notes[ii][0][1]
        else :
            temp.append(notes[ii][0])
            jj += 1

print ('bpm : ', tempo)
print (temp)

### end
```
