```python
#audio_load.py
import numpy as np
import scipy.io as sio
import scipy.io.wavfile
import matplotlib.pyplot as plt

rate, data = sio.wavfile.read('piano A.wav')
print ('rate: ', rate)
print ('data len: ', len(data))
print ('data', data) # sine 곡선을 이루고 있음.

size = len(data)
k = np.arange(size)
print ('k = ',k)

f0=k*rate/size    # double sides frequency range
print ('f0 = ',f0)

f0=f0[range(int(size/2))]
print ('f0 = ',f0)

Y=np.fft.fft(data)/size        # fft computing and normaliation
Y=Y[range(int(size/2))]

amplitude_Hz = 2*abs(Y)
print ('amplitude = ', amplitude_Hz)

plt.plot(f0,amplitude_Hz)
plt.show()

banks = np.where((amplitude_Hz >= (amplitude_Hz.max() * 0.90)) & (amplitude_Hz > 500))
print('banks =', banks)
pitchs = []
for bank in banks[0]:
    pitchs.append(f0[bank]) # f0에서 가져온다.
print('pitchs = ',pitchs)
print('f0[880,881]',f0[88285])
```
